```python
import numpy as np
import pandas as pd


url = (
    'https://storage.googleapis.com/esmartdata-courses-files/'
    'dash-course/data.csv'
)
df = pd.read_csv(url, index_col=0)
df.columns = [col.lower() for col in df.columns]
```

```python
df
```

```python
for i in range (len(df['power'])):
  df['power'][i] = np.where(df['power'][i] == 'null bhp', np.nan,df['power'][i])# df['power'][i] -> w przeciwnym razie pozostawiam wartosc tej kolumny to taki else :))
```
