```python
import numpy as np
import pandas as pd
import plotly.express as px

np.random.seed(42)
```

```python
X1 = np.array([1, 2, 3, 4, 5, 6])
Y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(X1)

print(f'Lata pracy: {X1}')
print(f'Wynagrodzenie: {Y}')
print(f'Liczba próbek: {m}')
```

```python
X1 = X1.reshape(m, 1)
Y = Y.reshape(-1, 1)
print(X1)
print(X1.shape)
```

```python
print(Y.shape)
print(Y)
```

```python
bias = np.ones((m, 1))
print(bias)
print(bias.shape)
```

```python
X = np.append(bias, X1, axis=1)
print(X)
print(X.shape)
```

```python
eta = 0.01 # wskaźnik uczenia, learning rate

#losowa inicjalizacja wag -> macierz 2 na 1
weights = np.random.randn(2, 1)
print(X)
#otrzymamy 2 punkty startowe w0 i w1
print(weights)
```

Gradient Prosty

```python
intercept = [] # w0
coef = [] #w1

for i in range(4000):
  gradient = (2 / m) * X.T.dot(X.dot(weights) - Y)
  weights = weights - eta * gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])
print(weights)
```

```python
df = pd.DataFrame(data={'intercept': intercept, 'coef': coef})
df.head()
```

```python
#w0
px.line(df, y='intercept', width=800, title='Dopasowanie: intercept')
```

```python
#w1
px.line(df, y='coef', width=800, title='Dopasowanie: coef')
```
