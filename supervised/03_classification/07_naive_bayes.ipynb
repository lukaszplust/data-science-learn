### <a name='0'></a> Import bibliotek

```
import numpy as np
import pandas as pd
import sklearn

np.random.seed(42)
np.set_printoptions(precision=6, suppress=True)
sklearn.__version__
```

### <a name='1'></a> Wygenerowanie danych

```
pogoda = ['słonecznie', 'deszczowo', 'pochmurno', 'deszczowo', 'słonecznie', 'słonecznie',
          'pochmurno', 'pochmurno', 'słonecznie']
temperatura = ['ciepło', 'zimno', 'ciepło', 'ciepło', 'ciepło', 'umiarkowanie',
               'umiarkowanie', 'ciepło', 'zimno']

spacer = ['tak', 'nie', 'tak', 'nie', 'tak', 'tak', 'nie', 'tak', 'nie']      

raw_df = pd.DataFrame(data={'pogoda': pogoda, 'temperatura': temperatura, 'spacer': spacer})
df = raw_df.copy()
df
```

### <a name='2'></a> Przygotowanie danych do modelu

```
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
df['spacer']= encoder.fit_transform(spacer)#na encoder wywołuje fit_transform(nazwa zmiennej)
df
```

```
df = pd.get_dummies(df, columns=['pogoda', 'temperatura'], drop_first=True)#get_dummies przypasowuje do wyszukiwanych(pogoda,temperatura) 0 oraz 1. Jesli dla pogoda_pochmurno jest pochmurno to 1, else 0 itd.
df
```

```
data = df.copy()
target = data.pop('spacer')

data
```

```
target#targetem sa 0 i 1 dla zmiennej spacer
```

```
data
```

### <a name='3'></a> Klasyfikator bayesowski

```
from sklearn.naive_bayes import GaussianNB

model = GaussianNB()
model.fit(data, target)
```

```
model.score(data, target)
```

```
data.iloc[[0]]#predykcja na 1 wierszu jest to slonecznie i cieplo
```
