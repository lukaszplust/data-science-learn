Macierz B -> wektor pobudzenia (o długości N)

```python
def wektor_pobudzenia(N):
  storage = []
  for i in range(N):
    storage.append(sin(i+6))
  return storage
```

```python
def norm(res):
    count = 0
    for i in range(len(res)):
        count += res[i] ** 2
    return math.sqrt(count)
```

Macierz A -> macierz systemowa

```python
#wytwarzam macierz pasmowa taka jak w pdf
def macierz_systemowa(a1, a2, a3, N):
    M = []
    for i in range(N):
        row = []
        for j in range(N):
            row.append(0)
        M.append(row)
    for i in range(N):#i to wiersz chyba xd
        for j in range(N):#j to kolumna chyba xd
            if i == j: #jesli ten sam wiersz co kolumna
                M[i][j] = a1 # to ustawiam a1 na glownej diagonali

                #to tutaj to chyba dolna macierz trojkatna 
                #hmm ten if z j-1 nie wiem czy jest konieczny
                if j - 1 >= 0:
                    M[i][j - 1] = a2
                    if j - 2 >= 0:
                        M[i][j - 2] = a3
                #to tutaj to chyba gorna macierz trojkatna 
                if j + 1 < N: #jesli kolumna + 1 jest mniejsza od maks rozmiaru to ustawiam to jako a2
                    M[i][j + 1] = a2
                    if j + 2 < N:#nastepnie ustawiam a3
                        M[i][j + 2] = a3
                break
    return M
```

```python
macierz_A = macierz_systemowa(a1, a2, a3, N)
macierz_B = wektor_pobudzenia(N)
```
